### 2.4.2 Tipado dinámico

En Python, el tipado es dinámico, lo que significa que no necesitas declarar explícitamente el tipo de una variable cuando la creas. El tipo de una variable se determina en tiempo de ejecución, según el valor que se le asigne. Esto hace que Python sea muy flexible y fácil de usar, especialmente para principiantes en programación.


Captura de Pantalla.png

```python
# Tipos de datos en Python

# Entero
edad = 30
print(type(edad))  # Salida: <class 'int'>

# Flotante
altura = 1.75
print(type(altura))  # Salida: <class 'float'>

# Cadena de texto
nombre = "Ariel"
print(type(nombre))  # Salida: <class 'str'>

# Booleano
es_estudiante = True
print(type(es_estudiante))  # Salida: <class 'bool'>

# Lista
hobbies = ["leer", "caminar", "cocinar"]
print(type(hobbies))  # Salida: <class 'list'>

# Cambio de tipo
edad = "treinta"
print(type(edad))  # Salida: <class 'str'>
```

## 2.5 Manipulación Avanzada de Cadenas

Además de concatenación y acceso por índice, Python ofrece muchas más operaciones y métodos para trabajar con cadenas. Considera agregar ejemplos de métodos como `split()`, `join()`, y `replace():`

- `split()` divide una cadena en una lista, basándose en un separador especificado.

```python
texto = "Python es increíble"
palabras = texto.split(" ")
print(palabras)
```

- `join()` une los elementos de una lista en una cadena, usando un separador especificado.

```python
palabras = ["Python", "es", "increíble"]
texto = " ".join(palabras)
print(texto)
```

- `replace()` reemplaza una subcadena por otra en la cadena.

```python
saludo = "Hola, mundo"
saludoNuevo = saludo.replace("mundo", "Python")
print(saludoNuevo)
```

## 2.6 Formateo de Cadenas

El formateo de cadenas es crucial para crear salidas legibles y formateadas dinámicamente. Python ofrece varias maneras de formatear cadenas, incluyendo el método `format()` y las `f-strings`:

- `Método format():`

```python
nombre = "Ariel"
mensaje = "Hola, {}. Bienvenido a Python.".format(nombre)
print(mensaje)
```

- `F-strings:`

```python
nombre = "Ariel"
mensaje = f"Hola, {nombre}. Bienvenido a Python."
print(mensaje)
```

## 2.7 Tuplas, Listas y Diccionarios

Python ofrece tres estructuras de datos para almacenar colecciones de datos de manera eficiente:

- **Listas**: Colecciones ordenadas y modificables de elementos.
- **Tuplas**: Colecciones ordenadas e inmutables de elementos.
- **Diccionarios**: Colecciones no ordenadas de pares clave-valor.

### 2.7.1 Tuplas
Las tuplas son colecciones inmutables, lo que significa que, una vez definidas, sus elementos no pueden ser modificados. Son ideales para almacenar datos que no cambiarán a lo largo de tu programa.

**Ejemplo de Tupla:**

Las tuplas son colecciones inmutables, lo que significa que, una vez definidas, sus elementos no pueden ser modificados. Son ideales para almacenar datos que no cambiarán a lo largo de tu programa.

```python
meses = ("Enero", "Febrero", "Marzo", "Abril", "Mayo", "Junio", "Julio", "Agosto", "Septiembre", "Octubre", "Noviembre", "Diciembre")
print(meses)
```

**Acceder a elementos:**

```python
dias_semana = ("Lunes", "Martes", "Miércoles", "Jueves", "Viernes")
print(dias_semana[2]) 
```

**Concatenación:**

```python
tupla1 = (1, 2, 3)
tupla2 = (4, 5, 6)
tupla3 = tupla1 + tupla2
print(tupla3)
```

### 2.7.2 Listas
A diferencia de las tuplas, las listas son colecciones que pueden modificarse, permitiéndote añadir, eliminar o cambiar elementos.

**Ejemplo de Lista:**

```python
libros_favoritos = ["Cien años de soledad", "El principito", "1984"]
print(libros_favoritos)
```

```python
libros_favoritos.append("El señor de los anillos")  # Añade un nuevo libro
print(libros_favoritos)
```

**Agregar elementos:**

```python
paises = ["México", "Argentina", "Brasil"]
paises.append("Colombia")  # Añade "Colombia" al final de la lista
print(paises)  
```

**Eliminar elementos:**

```python
paises = ["México", "Argentina", "Brasil", "Colombia"]
del paises[1]  # Elimina "Argentina" de la lista
print(paises) 
```

**Acceder a elementos:**

```python
paises = ["México", "Argentina", "Brasil", "Colombia"]
print(paises[2])  # Imprime "Brasil"
```

### 2.7.2 Diccionarios
Los diccionarios en Python son colecciones de pares clave-valor, lo que los hace ideales para almacenar datos relacionados, como un directorio telefónico.

**Ejemplo de Diccionario:**

```python
directorio = {"Juan Pérez": "555-0101", "Maria García": "555-0202"}
print(directorio)
```

```python
directorio["Luis Ramírez"] = "555-0303"  # Añade un nuevo contacto
print(directorio)
```

**Agregar pares clave-valor:**

```python
contactos = {"Juan": "1234-5678", "Ana": "8765-4321"}
contactos["Luis"] = "5555-6666"  # Añade un nuevo par clave-valor
print(contactos)
```

**Eliminar pares clave-valor:**

```python
contactos = {"Juan": "1234-5678", "Ana": "8765-4321", "Luis": "5555-6666"}
del contactos["Ana"]  # Elimina el par clave-valor correspondiente a "Ana"
print(contactos)
```

**Acceder a valores:**

```python
contactos = {"Juan": "1234-5678", "Ana": "8765-4321", "Luis": "5555-6666"}
print(contactos["Luis"])  # Imprime "5555-6666"
```

```python
print(contactos["Ana"][-1])
```
