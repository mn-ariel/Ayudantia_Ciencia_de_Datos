# 1. Pandas para Ciencia de Datos

Pandas es una librería poderosa en Python diseñada para el análisis y manipulación de datos. Ofrece estructuras de datos flexibles que facilitan el trabajo con datos tabulares y series temporales. Profundicemos en cómo podemos utilizar Pandas para tareas avanzadas en ciencia de datos.

## 1.2 Tipos de Datos en Pandas

Pandas maneja principalmente dos estructuras de datos:

- **Series**: Una serie es un array unidimensional capaz de almacenar cualquier tipo de dato, con un índice.
- **DataFram**e: Un DataFrame es una estructura de datos bidimensional, similar a una tabla de SQL o una hoja de cálculo de Excel, con columnas de diferentes tipos.


## 1.3 Creación de Series
Crear una serie en Pandas es sencillo. Puedes iniciar una serie desde listas, diccionarios, o arrays de NumPy:

````python
import pandas as pd

# Desde una lista
serie_lista = pd.Series([4, 7, -5, 3])
print(serie_lista)

# Desde un diccionario
serie_dict = pd.Series({'a': 4, 'b': 7, 'c': -3, 'd': 3})
print(serie_dict)
````

### 1.3.1 Atributos de Series

Las series en Pandas vienen con varios atributos útiles:

````python
print(serie_lista.values) # Muestra los valores
print(serie_lista.index)  # Muestra el índice de la serie
````

### 1.3.2 Acceso a Elementos de una Serie
Puedes acceder a los elementos de una serie mediante su índice:

````python
# Acceso por índice posicional
print(serie_lista[3])

# Acceso por índice de etiqueta (si es un diccionario)
print(serie_dict['a'])
````

## 1.4 **DataFrame**

Un objeto del tipo DataFrame define un conjunto de datos estructurado en forma de tabla donde cada columna es un objeto de tipo Series, es decir, todos los datos de una misma columna son del mismo tipo, y las filas son registros que pueden contender datos de distintos tipos.

Un DataFrame contiene dos índices, uno para las filas y otro para las columnas, y se puede acceder a sus elementos mediante los nombres de las filas y las columnas.

image.png

### 1.4.1 Creación de un DataFrame
Crear un DataFrame es tan simple como pasar un diccionario de listas o un array de NumPy a `pd.DataFrame()`:

````python
data = {'nombre':['María', 'Luis', 'Carmen', 'Antonio'],
      'edad':[18, 22, 20, 21], 
      'grado':['Economía', 'Medicina', 'Arquitectura', 'Economía'], 
      'correo':['maria@gmail.com', 'luis@yahoo.es', 'carmen@gmail.com', 'antonio@gmail.com']}
frame = pd.DataFrame(data)
print(frame)
````

### 1.4.2 Manipulación de DataFrames
Pandas proporciona numerosas herramientas para manipular DataFrames, incluyendo la reindexación, eliminación de columnas o filas, filtrado, etc.

````python
# Seleccionar una columna como serie
print(frame['edad'])

# Añadir una nueva columna
frame['deuda'] = 16.5

print(frame)
````

## 1.5 Funciones Esenciales en Pandas

- `.info()`: Proporciona un resumen conciso del DataFrame.
- `.describe()`: Ofrece estadísticas descriptivas de las columnas numéricas.
- `.value_counts()`: Cuenta la frecuencia de los valores únicos en una columna.
- `.groupby()`: Agrupa datos basados en alguna condición y permite aplicar funciones de agregación.

### 1.5.1 Acceso a los Elementos de un DataFrame
El acceso a los elementos se puede hacer por nombre de columna o por índice utilizando loc y iloc:

````python
# Creación de un DataFrame a partir de un diccionario
datos = {
    "Nombres": ["Juan", "Maria", "Leonardo", "Julieta", "Ana", "Luis", "Pedro", "Marta"],
    "Edades": [25, 30, 22, 29, 25, 30, 22, 29],
    "Ciudad": ["La Paz", "La Paz", "Cochabamba", "Santa Cruz", "Oruro", "Oruro", "Cochabamba", "Cochabamba"]
}
df = pd.DataFrame(datos)
print(df)
````

````python
# Acceso por índice
print(df.iloc[-1])
print("______________")
# Acceso por etiqueta
print(df.loc[7, 'Ciudad'])
````
