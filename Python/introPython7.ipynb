# 7. Manejo de Archivos en Python (File I/O) y Uso de Importaciones

## 7.1 Introducción al Manejo de Archivos

En Python, trabajar con archivos es una tarea común que permite la lectura, escritura y modificación de archivos externos. Esto es esencial para tareas como el procesamiento de datos, la lectura de configuraciones y el almacenamiento de información de manera persistente.

### 7.1.1 Abrir un Archivo

Para abrir un archivo en Python, utilizamos la función incorporada open(). Esta función requiere al menos un argumento: la ruta del archivo. Además, podemos especificar el modo en el que queremos abrir el archivo:

- **`'r'`**: Solo lectura.
- **`'w'`**: Solo escritura (sobrescribe el archivo).
- **`'a'`**: Añadir (escribe al final del archivo).
- **`'r+'`**: Lectura y escritura.

**Ejemplo de Apertura de Archivo para Lectura:**

```python
# Abrir un archivo de texto para lectura
with open('archivo.txt', 'r') as archivo:
    contenido = archivo.read()
    print(contenido)
```

Al usar el contexto `with open(...) as ...`, nos aseguramos de que el archivo se cierre automáticamente al final del bloque with, lo que es una buena práctica para manejar recursos.

**Leer y Escribir Archivos CSV**
Para trabajar con archivos CSV, Python ofrece el módulo csv.

```python
import csv

# Leer un archivo CSV
with open('datos.csv', mode='r') as archivo_csv:
    lector = csv.reader(archivo_csv)
    for fila in lector:
        print(fila)
```

```python
# Escribir un archivo CSV
with open('salida.csv', mode='w', newline='') as archivo_csv:
    escritor = csv.writer(archivo_csv)
    escritor.writerow(['nombre', 'edad'])
    escritor.writerow(['Juan', '30'])
```

```python
with open('salida.csv', mode='r') as archivo_csv:
    lector = csv.reader(archivo_csv)
    for fila in lector:
        print(fila)
```

### 7.1.2 Escritura Condicional a Archivos

Explora cómo se pueden realizar escrituras condicionales en archivos, útil para casos donde solo quieres añadir contenido bajo ciertas condiciones.

```python
with open('log.txt', 'a') as archivo_log:
    evento = 'Error de red'
    if evento == 'Error de red':
        archivo_log.write('Se detectó un error de red\n')
```

### 7.1.3 Trabajar con Archivos JSON

El formato JSON es ampliamente utilizado en el intercambio de datos en aplicaciones web. Mostrar cómo Python puede manejar archivos JSON para leer y escribir datos estructurados puede ser muy útil.

```python
import json

# Escribir un archivo JSON
datos = {"nombre": "Marcelo", "edad": 30}
with open('datos.json', 'w') as archivo_json:
    json.dump(datos, archivo_json)
```

```python
# Leer un archivo JSON
with open('datos.json', 'r') as archivo_json:
    datos = json.load(archivo_json)
    print(datos)
```

### 7.1.4 Operaciones con Archivos Binarios

A veces, es necesario trabajar con archivos binarios directamente, por ejemplo, para la manipulación de imágenes o para leer datos que no están en formato de texto.

```python
# Leer un archivo binario
with open('python.png', 'rb') as archivo_binario:
    contenido = archivo_binario.read()
    print(contenido[:20])  # Muestra los primeros 20 bytes
```

**Ejemplo usó de la biblioteca Math**

La biblioteca math proporciona acceso a funciones matemáticas.

```python
import math

print(math.pi)

resultado = math.sqrt(16)
print(resultado)
```

**Importar de Forma Selectiva**

Podemos importar funciones o constantes específicas de un módulo para usarlas directamente sin necesidad de prefijarlas con el nombre del módulo.

```python
from math import pi, sqrt

area_circulo = pi * sqrt(5)
print(area_circulo)
```
